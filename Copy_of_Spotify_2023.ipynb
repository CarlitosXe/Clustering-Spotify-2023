```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

# Load hasil PCA dari file
pca_result = np.load('pca_result.npy')  # Load file .npy

# Membuat model KMeans dengan 3 klaster dan melakukan fit
kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')
kmeans.fit(pca_result)

# Mendapatkan label klaster
labels = kmeans.labels_  # Label klaster untuk setiap data

# Hitung silhouette score
sil_score = silhouette_score(pca_result, labels)
print(f"Silhouette Score untuk 3 klaster: {sil_score:.4f}")
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import numpy as np

# Assuming 'df_cleaned' and 'features_for_clustering' are defined as before
# ... (Previous code to define df_cleaned and features_for_clustering) ...

# Define features_for_clustering here, assuming 'df_cleaned' exists
# Replace 'relevant_columns' with the actual columns you want to use
features_for_clustering = dataset[['danceability_%', 'energy_%', 'valence_%', 'acousticness_%',
                                      'instrumentalness_%']]

# Standardize the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features_for_clustering) # This line defines scaled_features


# Drop rows with NaN values from the scaled features before applying DBSCAN
scaled_features_no_nan = pd.DataFrame(scaled_features).dropna().values

# Initialize variables to track the best DBSCAN score and corresponding parameters
best_dbscan_score = -1
best_eps = None
best_min_samples = None

# Store the results for plotting or comparison
dbscan_scores = []

# Iterate over a range of eps and min_samples values to fine-tune DBSCAN parameters
for eps in np.arange(0.1, 1.1, 0.1):  # Test eps values from 0.1 to 1.0
    for min_samples in range(2, 10):  # Test min_samples from 2 to 9
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        dbscan_labels = dbscan.fit_predict(scaled_features_no_nan)

        # Check if there are at least two clusters (excluding noise)
        if len(set(dbscan_labels)) > 1:
            score = silhouette_score(scaled_features_no_nan, dbscan_labels)
            dbscan_scores.append((eps, min_samples, score))

            # Update best score and parameters if current score is better
            if score > best_dbscan_score:
                best_dbscan_score = score
                best_eps = eps
                best_min_samples = min_samples

# Print the best DBSCAN parameters and corresponding Silhouette Score
print(f'Best DBSCAN Silhouette Score: {best_dbscan_score:.3f} with eps={best_eps} and min_samples={best_min_samples}')

# Convert the results to a DataFrame
dbscan_scores_df = pd.DataFrame(dbscan_scores, columns=['eps', 'min_samples', 'silhouette_score'])

# Use pivot_table instead of pivot for the heatmap
pivot = dbscan_scores_df.pivot_table(values='silhouette_score', index='eps', columns='min_samples')

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot, annot=True, fmt='.3f', cmap='coolwarm')
plt.title('DBSCAN Silhouette Scores for Different eps and min_samples')
plt.xlabel('min_samples')
plt.ylabel('eps')
plt.show()
```
